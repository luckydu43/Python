```python
np.random.seed(0)
m=100
X = np.linspace(0, 10, m).reshape(m,1)
y = X + np.random.randn(m,1)
plt.scatter(X, y)
```

```python
lr = linear_model.LinearRegression()
lr.fit(X,y)
```

```python
lr.score(X,y)
```

```python
predictions = lr.predict(X)
plt.scatter(X,y)
plt.scatter(X,predictions, c='r')
```
